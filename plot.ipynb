```python
import matplotlib.pyplot as plt
```

```python
processors = [2**i for i in range(2,8)]
min_runtime = [0.0854381,
0.052956,
0.0245038,
0.178231,
0.102576,
0.0969837,
]
avg_runtime = [0.0862865,
0.0532694,
0.0248831,
0.187266,
0.111109,
0.12649,
]
max_runtime = [0.0873831,
0.0537113,
0.0254102,
0.188924,
0.112268,
0.128056,
]
plt.plot(processors,min_runtime)
plt.plot(processors,avg_runtime)
plt.plot(processors,max_runtime)
plt.xlabel("# of Processors")
plt.ylabel("Runtime (s)")
plt.title('Execution times for running on a 512x512 board for 500 iterations')
plt.legend(['min','avg','max'])
plt.plot()
```
